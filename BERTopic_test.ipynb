## 1.3 資料及驗證

````python
def validate_dataset(dataset):
    """
    驗證資料集格式和內容
    """
    print("=== 資料集驗證 ===")
    print(f"資料筆數: {len(dataset)}")
    print(f"欄位: {list(dataset.features.keys())}")
    
    # 檢查前幾筆資料
    print("\n前3筆資料範例:")
    for i in range(min(3, len(dataset))):
        print(f"\n第 {i+1} 筆:")
        print(f"標題: {dataset[i]['title'][:50]}...")
        print(f"內容: {dataset[i]['transcript'][:100]}...")
    
    # 統計資訊
    transcript = dataset['transcript']
    word_counts = [len(text.split()) for text in transcript[:100]]  # 取樣100筆
    print(f"\n平均字數: {sum(word_counts) / len(word_counts):.1f}")
````

## 2.2 清理其中的廠商、模型名稱

### 定義清理資料的函式

````python
def remove_brands_case_sensitive(text):
    """
    移除品牌關鍵字（大小寫敏感，只移除大寫版本）
    這個方法比較安全，因為一般品牌名都是大寫開頭
    """
    BRAND_KEYWORDS = ['Bentley Motors', 'Lucid', 'Scion', 'Cadillac', 'Hyundai', 'Mercedes-Benz', 'BYD', 
                      'Acura', 'Saturn', 'Lexus', 'Nissan', 'Karma', 'Mazda', 'Rolls-Royce',
                      'Wheego Electric Cars, Inc.', 'McLaren', 'Genesis', 'Chrysler', 'Volkswagen', 'Coda Automotive',
                      'Fiat', 'Infiniti', 'Buick', 'Chevrolet', 'Kandi', 'Lincoln', 'Porsche', 'Tesla', 'Mitsubishi', 
                      'Audi', 'Volvo', 'Honda', 'BMW', 'Dodge', 'Ford', 'Maserati', 'Toyota', 'Fisker Automotive',
                      'Ferrari', 'Kia', 'Mercury', 'Bugatti Rimac', 'Polestar']
    
    pattern = r"\b(" + "|".join(re.escape(word) for word in BRAND_KEYWORDS) + r")\b"
    return re.sub(pattern, "", text)
````

````python
def clean_text_improved(text):
    """
    改進的文本清理函數
    """
    # 選擇其中一種品牌移除方法
    text = remove_brands_case_sensitive(text)  # 或使用 remove_brands_whole_word(text)
    
    # 移除多餘空白
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()
````

### 應用清洗，再送入模型

````python
import re
sentences = [clean_text_improved(s) for s in sentences]
````

# 3️⃣ 模型預先設定

## 3.1 預先計算嵌入

在創建了我們的數據（即摘要 ）之後，我們可以深入研究第一個最佳實踐， 即預先計算嵌入 。

BERTopic 的工作原理是將文檔轉換為稱為嵌入的數值。這個過程可能非常昂貴，特別是當我們想反覆運算參數時。相反，我們可以計算一次這些嵌入，然後將它們提供給 BERTopic 以跳過每次計算嵌入。

````python
from sentence_transformers import SentenceTransformer

# Pre-calculate embeddings
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = embedding_model.encode(sentences, show_progress_bar=True)
````

## 3.2 防止隨機行為

在 BERTopic 中，我們通常使用降維演算法來減小嵌入的大小。這樣做是為了在一定程度上防止維度的詛咒 。

默認情況下，這是通過 UMAP 完成的，UMAP 是一種用於減少維度空間的不可思議的演算法。但是，默認情況下，它顯示隨機行為，每次運行它時都會產生不同的結果。為了防止這種情況，我們需要在將模型傳遞給 BERTopic 之前設置模型的 random_state。

因此，我們現在可以在每次運行模型時完全重現結果。

````python
from umap import UMAP

umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric='cosine', random_state=42)
````

## 3.3 控制 Topic 的數量

有一個參數來控制 Topic 的數量，即 nr_topics。但是，此參數會在創建主題後合併主題。它是一個支援創建固定數量主題的參數。

但是，建議通過集群模型控制主題的數量，默認情況下是 HDBSCAN。HDBSCAN 有一個參數，即 min_topic_size 間接控制將要創建的主題數量。

較高的 min_topic_size 將生成較少的主題，較低的 min_topic_size 將生成更多的主題。

在這裡，我們將使用 min_topic_size=40 來繞過 XXX 主題。

````python
from hdbscan import HDBSCAN

hdbscan_model = HDBSCAN(min_cluster_size=150, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
````

````python
from hdbscan import HDBSCAN
import numpy as np

from umap import UMAP

# 降到 5 維以內即可（通常 2~5）
reduced_embeddings = UMAP(n_components=5, random_state=42).fit_transform(embeddings)

for size in [50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400, 500, 1000]:
    clusterer = HDBSCAN(min_cluster_size=size, prediction_data=True)
    cluster_labels = clusterer.fit_predict(reduced_embeddings)

    n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
    noise_ratio = np.mean(cluster_labels == -1)

    print(f"min_cluster_size={size}, n_clusters={n_clusters}, noise_ratio={noise_ratio:.2%}")
````

## 3.4 改進預設表示

主題的預設表示形式是通過 c-TF-IDF 計算的。但是，c-TF-IDF 由 CountVectorizer 提供支援，該 CountVectorizer 可將文本轉換為標記。使用 CountVectorizer，我們可以做很多事情：

- 刪除停用詞
- 忽略不常見的單詞
- 增加

換句話說，我們可以在將文檔分配給主題后預處理主題表示。這不會以任何方式影響聚類過程。

在這裡，我們將忽略英語停用詞和不常見的單詞。此外，通過增加 n-gram 範圍，我們將考慮由一個或兩個單詞組成的主題表示。

````python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english", min_df=2, ngram_range=(1, 2))
````

## 3.5 其他表示

以前，我們已經調整了預設表示形式，但 BERTopic 中還有相當多的其他主題表示形式可供我們選擇。從 KeyBERTInspired 和 PartOfSpeech，到 OpenAI 的 ChatGPT 和開源替代品，許多表示都是可能的。

在 BERTopic 中，您可以同時對許多不同的主題表示進行建模，以測試它們並獲得主題描述的不同視角。這稱為多方面主題建模。

在這裡，我們將演示 BERTopic 中許多有趣且有用的表示形式：

````python
import openai
from openai import OpenAI
from bertopic.representation import KeyBERTInspired, MaximalMarginalRelevance, OpenAI, PartOfSpeech

# KeyBERT
keybert_model = KeyBERTInspired()

# Part-of-Speech
pos_model = PartOfSpeech("en_core_web_sm")

# MMR
mmr_model = MaximalMarginalRelevance(diversity=0.3)

# gpt-4o-mini
prompt = """
I have a topic that contains the following documents:
[DOCUMENTS]
The topic is described by the following keywords: [KEYWORDS]

Based on the information above, extract a short but highly descriptive topic label of at most 5 words. Make sure it is in the following format:
topic: <topic label>
"""
client = openai.OpenAI(api_key="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")
openai_model = OpenAI(client, model="gpt-4o-mini", exponential_backoff=True, chat=True, prompt=prompt)

# All representation models
representation_model = {
    "KeyBERT": keybert_model,
    "OpenAI": openai_model,
    "MMR": mmr_model,
    "POS": pos_model
}
````
